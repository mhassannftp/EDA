## Destructive Features or Most Occurning

```python
print("Country with the most attact = ",df['Country'].value_counts().idxmax())
print("Region with the most attact = ",df['Region'].value_counts().idxmax())
print("City with the most attact = ",df['city'].value_counts().index[1])
print("Year with the most attact = ",df['Year'].value_counts().idxmax())
print("Month with the most attact = ",df['Month'].value_counts().idxmax())
print("Group with the most attact = ",df['Group'].value_counts().index[1])
print("Mostattact Type = ",df['AttackType'].value_counts().idxmax())
```

## Terrorist Activities by Region in each Year.

```python
pd.crosstab(df.Year, df.Region).plot(kind='area', figsize=(15,10))
plt.title("Terrorist Activities by Region in each Year")
plt.ylabel("Number of Attacts")
plt.show()
```

## Scicide Attack by Region in each Year

```python
pd.crosstab(df.Region, df.suicide).plot(kind='bar', figsize=(12,8))
plt.title("Suicide Attact by Region in each Year")
plt.ylabel("Number of Attacts")
plt.show();
```

## Top 20 Countries affected by Terror Attacks

```python
df.Country.value_counts()[:20]
```

```python
plt.subplots(figsize=(18,8))
sns.color_palette("viridis")
sns.barplot(df['Country'].value_counts()[:20].index,df['Country'].value_counts()[:20].values)
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation= 90)
plt.show()
```

Iraq is the most affacted Country and second most affected country is Pakistan
